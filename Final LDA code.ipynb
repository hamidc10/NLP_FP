```python
training_data_outdoors = []
training_data_kitchen=[]
training_data_books = []
training_data_electronics = []

with open('Home_and_Kitchen_5.json') as f:
    for l in f:
        training_data_kitchen.append(json.loads(l.strip()))
print("added_home_lst")
with open('Sports_and_Outdoors_5.json') as f:
    for l in f:
        training_data_outdoors.append(json.loads(l.strip()))
print("added_sports_lst")      
with open('Books_5.json') as f:
    for l in f:
        training_data_books.append(json.loads(l.strip()))
print("added_books_lst") 
with open('Electronics_5.json') as f:
    for l in f:
        training_data_electronics.append(json.loads(l.strip()))
print("added_Electronics_lst")
```

```python
# Used this for refernce to help with LDA:
#1.https://towardsdatascience.com/end-to-end-topic-modeling-in-python-latent-dirichlet-allocation-lda-35ce4ed6b3e0
#2.https://neptune.ai/blog/pyldavis-topic-modelling-exploration-tool-that-every-nlp-data-scientist-should-know

#Here it cleans the list into one big complete list then cleans the data to use for the LDA model
# Here is the chatgpt link for the workers: https://chat.openai.com/share/ba8e7fa6-3590-4209-b97f-b32f6645e7d1
# Here is the orginal promt:"how long do you think it would take to train 90% of a 22gb file using LDA"
```

```python
complete_test=[]
#Differnt test I tried:
#make the list (:int(0.8*len("name of data set")))
#right now set each to 100k but can run at least I have run the complete kicthen,sports, and books up to 1 million
complete_test.extend(training_data_outdoors[:100000])
complete_test.extend(training_data_kitchen[:100000])
complete_test.extend(training_data_books[:100000])
complete_test.extend(training_data_electronics[:100000])


simplified_data=[]
for i in range(len(complete_test)):
    simplified_data.append(complete_test[i].get("summary",'')+' '+complete_test[i].get("reviewText",''))
#Here for debugging:
#print(len(simplified_data))

def sent_to_words(data):
    for obj in data:
        yield(gensim.utils.simple_preprocess(str(obj), deacc=True))  # deacc=True removes punctuations


def stopwordscleaner(data):
    stp = set(stopwords.words('english'))
    return [[tok for tok in obj if tok.lower() not in stp] for obj in data]
    
clean_words=sent_to_words(simplified_data)
clean_data=stopwordscleaner(clean_words)


id2word=corpora.Dictionary(clean_data)
reviews=clean_data
review_corpus=[id2word.doc2bow(review) for review in reviews]

#Here is where you change the number of topics
topic=100
# Also make sure to change the number of workers based on number of CPU'S
lda = gensim.models.ldamulticore.LdaMulticore(corpus=review_corpus,id2word=id2word,num_topics=topic,workers=20)
# pprint(lda.print_topics())

LDAvis_prepared = pyLDAvis.gensim.prepare(lda, review_corpus, id2word)
pyLDAvis.save_html(LDAvis_prepared,'lda_model'+'.html')
LDAvis_prepared
```
